```python
class Robot:
    def __init__(self, name, x=0, y=0):
        self.name = name
        self.x = x
        self.y = y
        self.energy = 100
        self.history = [(x, y)]

    def move_up(self):
        self.y += 1
        self.history.append((self.x, self.y))
        self.energy -= 10

    def move_down(self):
        self.y -= 1
        self.history.append((self.x, self.y))
        self.energy -= 10

    def get_max_coordinate(self):
        return max(abs(self.x), abs(self.y))

    def get_quadrant(self):
        if self.x == 0 and self.y == 0:
            return "–¶–µ–Ω—Ç—Ä"
        elif self.x == 0 or self.y == 0:
            return "–û—Å—å"
        elif self.x > 0 and self.y > 0:
            return 1
        elif self.x < 0 and self.y > 0:
            return 2
        elif self.x < 0 and self.y < 0:
            return 3
        else:
            return 4

    def predict_energy_after_moves(self, commands_list):
        return self.energy - len(commands_list) * 10
```

```python
class Robot:
    def __init__(self, name):
        self.name = name

    def add_item(self, item_name, quantity=1):
        item_attr = f"item_{item_name}"
        if hasattr(self, item_attr):
            setattr(self, item_attr, getattr(self, item_attr) + quantity)
        else:
            setattr(self, item_attr, quantity)

    def remove_item(self, item_name, quantity=1):
        item_attr = f"item_{item_name}"
        if hasattr(self, item_attr):
            current_quantity = getattr(self, item_attr)
            if current_quantity <= quantity:
                delattr(self, item_attr)
            else:
                setattr(self, item_attr, current_quantity - quantity)

    def get_item_count(self, item_name):
        item_attr = f"item_{item_name}"
        return getattr(self, item_attr, 0)

    def show_inventory(self):
        print(f"üì¶ –ò–Ω–≤–µ–Ω—Ç–∞—Ä—å {self.name}:")
        found_items = False
        for attr in dir(self):
            if attr.startswith("item_"):
                print(f"{attr[5:]}: {getattr(self, attr)}")
                found_items = True
        if not found_items:
            print("–ü—É—Å—Ç–æ")

    def get_total_items(self):
        total_count = 0
        for attr in dir(self):
            if attr.startswith("item_"):
                total_count += getattr(self, attr)
        return total_count
```

```python
class Robot:
    def __init__(self, name, x=0, y=0):
        self.name = name
        self.x = x
        self.y = y
        self._access_level = 1
        self._security_log = []
        self._failed_attempts = 0
        self._blocked_operations = 0

    def _log_security_event(self, status, action, details=""):
        entry = f"{status}: {action}"
        if details:
            entry += f" ({details})"
        self._security_log.append(entry)

    def _check_access(self, required_level, action_name):
        if self._blocked_operations > 0:
            self._blocked_operations -= 1
            self._log_security_event("BLOCKED", action_name)
            return False
        if self._access_level >= required_level:
            self._failed_attempts = 0
            return True
        else:
            self._failed_attempts += 1
            self._log_security_event("DENIED", action_name, f"access level {self._access_level} < {required_level}")
            if self._failed_attempts >= 3:
                self._blocked_operations = 5
                self._failed_attempts = 0
            return False

    def set_access_level(self, level, password):
        if password == "admin123":
            self._access_level = level
            self._log_security_event("SUCCESS", "set_access_level")
        else:
            self._failed_attempts += 1
            self._log_security_event("DENIED", "set_access_level", "wrong password")
            if self._failed_attempts >= 3:
                self._blocked_operations = 5
                self._failed_attempts = 0

    def secure_teleport(self, x_new, y_new):
        if self._check_access(3, "secure_teleport"):
            self.x = x_new
            self.y = y_new
            self._log_security_event("SUCCESS", "secure_teleport")

    def emergency_shutdown(self):
        if self._check_access(5, "emergency_shutdown"):
            self._log_security_event("SUCCESS", "emergency_shutdown")

    def view_security_log(self):
        if self._check_access(2, "view_security_log"):
            for entry in self._security_log:
                print(entry)
            self._log_security_event("SUCCESS", "view_security_log")
```

```python
import unittest
from IPython.display import display, HTML

class TestRobotSecurity(unittest.TestCase):
    def setUp(self):
        self.robot = Robot("–¢–µ—Å—Ç–æ–≤—ã–π")
        self.robot._security_log = []  # –û—á–∏—â–∞–µ–º –ª–æ–≥ –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ç–µ—Å—Ç–∞
        self.robot._failed_attempts = 0
        self.robot._blocked_operations = 0
        self.robot._access_level = 1

    def test_set_access_level_success(self):
        self.robot.set_access_level(3, "admin123")
        self.assertEqual(self.robot._access_level, 3)
        self.assertIn("SUCCESS: set_access_level", self.robot._security_log[0])

    def test_set_access_level_fail_wrong_password(self):
        self.robot.set_access_level(3, "wrongpass")
        self.assertEqual(self.robot._access_level, 1)
        self.assertIn("DENIED: set_access_level", self.robot._security_log[0])

    def test_secure_teleport_denied_access(self):
        self.robot.secure_teleport(10, 10)
        self.assertEqual(self.robot.x, 0)
        self.assertIn("DENIED: secure_teleport (access level 1 < 3)", self.robot._security_log[0])

    def test_secure_teleport_success(self):
        self.robot.set_access_level(3, "admin123")
        self.robot.secure_teleport(10, 10)
        self.assertEqual(self.robot.x, 10)
        self.assertEqual(self.robot.y, 10)
        self.assertIn("SUCCESS: secure_teleport", self.robot._security_log[1])

    def test_emergency_shutdown_denied_access(self):
        self.robot.emergency_shutdown()
        self.assertIn("DENIED: emergency_shutdown (access level 1 < 5)", self.robot._security_log[0])

    def test_emergency_shutdown_success(self):
        self.robot.set_access_level(5, "admin123")
        self.robot.emergency_shutdown()
        self.assertIn("SUCCESS: emergency_shutdown", self.robot._security_log[1])

    def test_view_security_log_denied_access(self):
        # Log should contain the denial itself
        self.robot.view_security_log()
        self.assertIn("DENIED: view_security_log (access level 1 < 2)", self.robot._security_log[0])

    def test_view_security_log_success(self):
        self.robot.set_access_level(2, "admin123")
        self.robot.view_security_log()
        self.assertGreater(len(self.robot._security_log), 0)
        self.assertIn("SUCCESS: view_security_log", self.robot._security_log[1])

    def test_failed_attempts_blocking(self):
        self.robot.set_access_level(0, "wrongpass") # 1
        self.robot.set_access_level(0, "wrongpass") # 2
        self.robot.set_access_level(0, "wrongpass") # 3 - should block

        self.assertTrue(self.robot._blocked_operations > 0)
        # –°–ª–µ–¥—É—é—â–∞—è –ø–æ–ø—ã—Ç–∫–∞ –¥–æ–ª–∂–Ω–∞ –±—ã—Ç—å –∑–∞–±–ª–æ–∫–∏—Ä–æ–≤–∞–Ω–∞
        self.robot.secure_teleport(1, 1)
        self.assertIn("BLOCKED: secure_teleport", self.robot._security_log[-1])
        self.assertEqual(self.robot.x, 0) # –ù–µ –¥–æ–ª–∂–Ω–æ –±—ã–ª–æ —Å–¥–≤–∏–Ω—É—Ç—å—Å—è

    def test_blocked_operations_countdown(self):
        # Block the robot first
        for _ in range(3):
            self.robot.set_access_level(0, "wrongpass")

        self.robot.secure_teleport(1, 1) # Blocked (1/5)
        self.robot.secure_teleport(1, 1) # Blocked (2/5)
        self.robot.secure_teleport(1, 1) # Blocked (3/5)
        self.robot.secure_teleport(1, 1) # Blocked (4/5)
        self.robot.secure_teleport(1, 1) # Blocked (5/5)

        # Now it should be unblocked
        self.robot.set_access_level(3, "admin123")
        self.robot.secure_teleport(1, 1) # Should succeed
        self.assertEqual(self.robot.x, 1)
        self.assertIn("SUCCESS: secure_teleport", self.robot._security_log[-1])

# –ó–∞–ø—É—Å–∫ —Ç–µ—Å—Ç–æ–≤
if __name__ == '__main__':
    suite = unittest.TestSuite()
    suite.addTest(unittest.makeSuite(TestRobotSecurity))
    runner = unittest.TextTestRunner(verbosity=2)

    # –ó–∞—Ö–≤–∞—Ç—ã–≤–∞–µ–º –≤—ã–≤–æ–¥, —á—Ç–æ–±—ã –æ–Ω –Ω–µ –º–µ—à–∞–ª —Ñ–æ—Ä–º–∞—Ç–∏—Ä–æ–≤–∞–Ω–∏—é Jupyter
    import io
    from contextlib import redirect_stdout

    f = io.StringIO()
    with redirect_stdout(f):
        runner.run(suite)

    s = f.getvalue()

    # –§–æ—Ä–º–∞—Ç–∏—Ä—É–µ–º –≤—ã–≤–æ–¥ –¥–ª—è Jupyter
    html_output = ""
    for line in s.splitlines():
        if "PASSED" in line:
            html_output += f"{line}"
        elif "FAILED" in line or "ERROR" in line:
            html_output += f"{line}"
        else:
            html_output += f"{line}"
    html_output += ""

    display(HTML(html_output))
```

```python
import random

class NullBuilderRobot:
    def __getattr__(self, name):
        return lambda *args, **kwargs: self

_null_builder_robot = NullBuilderRobot()

class BuilderRobot:
    def __init__(self, name):
        self.name = name
        self.current_building = None
        self.buildings_completed = []
        self.total_cost = 0

    def start_building(self, building_name):
        print(f"üèóÔ∏è {self.name} –Ω–∞—á–∏–Ω–∞–µ—Ç —Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–æ: {building_name}")
        self.current_building = {
            "name": building_name,
            "material": "–Ω–µ–∏–∑–≤–µ—Å—Ç–Ω–æ",
            "floors": {"–æ–±—ã—á–Ω—ã–π": []},
            "windows": 0,
            "foundation_laid": False,
            "walls_built": False,
            "roof_installed": False,
            "quality": "–Ω–µ –ø—Ä–æ–≤–µ—Ä–µ–Ω–æ"
        }
        return self

    def add_floor(self, floor_type="–æ–±—ã—á–Ω—ã–π"):
        if not self.current_building:
            print("‚ùå –°–Ω–∞—á–∞–ª–∞ –Ω–∞—á–Ω–∏—Ç–µ —Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–æ –∑–¥–∞–Ω–∏—è!")
            return self
        if floor_type not in self.current_building["floors"]:
            self.current_building["floors"][floor_type] = []
        print(f"‚ûï –î–æ–±–∞–≤–ª–µ–Ω —ç—Ç–∞–∂: {floor_type}")
        return self

    def add_room(self, room_name, floor_number=1):
        if not self.current_building:
            print("‚ùå –°–Ω–∞—á–∞–ª–∞ –Ω–∞—á–Ω–∏—Ç–µ —Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–æ –∑–¥–∞–Ω–∏—è!")
            return self
        floor_types = list(self.current_building["floors"].keys())
        if floor_number > 0 and floor_number <= len(floor_types):
            floor_key = floor_types[floor_number - 1]
            self.current_building["floors"][floor_key].append(room_name)
            print(f"üö™ –î–æ–±–∞–≤–ª–µ–Ω–∞ –∫–æ–º–Ω–∞—Ç–∞ '{room_name}' –Ω–∞ —ç—Ç–∞–∂ '{floor_key}'")
        else:
            print(f"‚ùå –ù–µ —É–¥–∞–ª–æ—Å—å –¥–æ–±–∞–≤–∏—Ç—å –∫–æ–º–Ω–∞—Ç—É: —ç—Ç–∞–∂ '{floor_number}' –Ω–µ –Ω–∞–π–¥–µ–Ω.")
        return self

    def set_material(self, material):
        if not self.current_building:
            print("‚ùå –°–Ω–∞—á–∞–ª–∞ –Ω–∞—á–Ω–∏—Ç–µ —Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–æ –∑–¥–∞–Ω–∏—è!")
            return self
        self.current_building["material"] = material
        print(f"üß± –£—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω –º–∞—Ç–µ—Ä–∏–∞–ª: {material}")
        return self

    def lay_foundation(self):
        if not self.current_building:
            print("‚ùå –°–Ω–∞—á–∞–ª–∞ –Ω–∞—á–Ω–∏—Ç–µ —Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–æ –∑–¥–∞–Ω–∏—è!")
            return self
        self.current_building["foundation_laid"] = True
        print(f"üèóÔ∏è –ó–∞–ª–æ–∂–µ–Ω —Ñ—É–Ω–¥–∞–º–µ–Ω—Ç –¥–ª—è {self.current_building['name']}")
        return self

    def build_walls(self):
        if not self.current_building or not self.current_building["foundation_laid"]:
            print("‚ùå –°–Ω–∞—á–∞–ª–∞ –∑–∞–ª–æ–∂–∏—Ç–µ —Ñ—É–Ω–¥–∞–º–µ–Ω—Ç!")
            return self
        self.current_building["walls_built"] = True
        print(f"üèóÔ∏è –ü–æ—Å—Ç—Ä–æ–µ–Ω—ã —Å—Ç–µ–Ω—ã –¥–ª—è {self.current_building['name']}")
        return self

    def install_roof(self):
        if not self.current_building or not self.current_building["walls_built"]:
            print("‚ùå –°–Ω–∞—á–∞–ª–∞ –ø–æ—Å—Ç—Ä–æ–π—Ç–µ —Å—Ç–µ–Ω—ã!")
            return self
        self.current_building["roof_installed"] = True
        print(f"üèóÔ∏è –£—Å—Ç–∞–Ω–æ–≤–ª–µ–Ω–∞ –∫—Ä—ã—à–∞ –¥–ª—è {self.current_building['name']}")
        return self

    def add_windows(self, count):
        if not self.current_building or not self.current_building["walls_built"]:
            print("‚ùå –°–Ω–∞—á–∞–ª–∞ –ø–æ—Å—Ç—Ä–æ–π—Ç–µ —Å—Ç–µ–Ω—ã, —á—Ç–æ–±—ã –¥–æ–±–∞–≤–∏—Ç—å –æ–∫–Ω–∞!")
            return self
        self.current_building["windows"] = count
        print(f"üñºÔ∏è –î–æ–±–∞–≤–ª–µ–Ω–æ {count} –æ–∫–æ–Ω")
        return self

    def inspect_quality(self):
        if not self.current_building:
            print("‚ùå –°–Ω–∞—á–∞–ª–∞ –Ω–∞—á–Ω–∏—Ç–µ —Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–æ –∑–¥–∞–Ω–∏—è!")
            return self
        quality_good = random.choice([True, False])
        self.current_building["quality"] = "—Ö–æ—Ä–æ—à–µ–µ" if quality_good else "–ø–ª–æ—Ö–æ–µ"
        print(f"üîç –ü—Ä–æ–≤–µ—Ä–µ–Ω–æ –∫–∞—á–µ—Å—Ç–≤–æ: {self.current_building['quality']}")
        return self

    def if_quality_good(self):
        if not self.current_building:
            return _null_builder_robot
        if self.current_building["quality"] == "—Ö–æ—Ä–æ—à–µ–µ":
            return self
        else:
            print(f"‚ùå –ö–∞—á–µ—Å—Ç–≤–æ –∑–¥–∞–Ω–∏—è {self.current_building['name']} –Ω–µ—É–¥–æ–≤–ª–µ—Ç–≤–æ—Ä–∏—Ç–µ–ª—å–Ω–æ–µ. –ü—Ä–æ–ø—É—Å–∫–∞–µ–º –¥–∞–ª—å–Ω–µ–π—à–∏–µ –¥–µ–π—Å—Ç–≤–∏—è.")
            return _null_builder_robot

    def calculate_cost(self):
        if not self.current_building:
            print("‚ùå –°–Ω–∞—á–∞–ª–∞ –Ω–∞—á–Ω–∏—Ç–µ —Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–æ –∑–¥–∞–Ω–∏—è!")
            return self
        cost = 0
        cost += 1000 if self.current_building["foundation_laid"] else 0
        cost += 2000 if self.current_building["walls_built"] else 0
        cost += 1500 if self.current_building["roof_installed"] else 0
        cost += self.current_building["windows"] * 100
        for floor_type, rooms in self.current_building["floors"].items():
            cost += 500
            cost += len(rooms) * 200
        material_cost = {"–∫–∏—Ä–ø–∏—á": 500, "–±–µ—Ç–æ–Ω": 400, "–¥–µ—Ä–µ–≤–æ": 300, "–∂–µ–ª–µ–∑–æ–±–µ—Ç–æ–Ω": 600}.get(self.current_building["material"], 0)
        cost += material_cost
        self.total_cost += cost
        print(f"üí∞ –†–∞—Å—Å—á–∏—Ç–∞–Ω–∞ —Å—Ç–æ–∏–º–æ—Å—Ç—å –¥–ª—è {self.current_building['name']}: {cost} (–æ–±—â–∞—è: {self.total_cost})")
        return self

    def finish_building(self):
        if not self.current_building:
            print("‚ùå –ù–µ—Ç —Ç–µ–∫—É—â–µ–≥–æ –∑–¥–∞–Ω–∏—è –¥–ª—è –∑–∞–≤–µ—Ä—à–µ–Ω–∏—è!")
            return self
        if not all([self.current_building["foundation_laid"], self.current_building["walls_built"], self.current_building["roof_installed"]]):
            print(f"‚ö†Ô∏è –ó–¥–∞–Ω–∏–µ {self.current_building['name']} –Ω–µ –ø–æ–ª–Ω–æ—Å—Ç—å—é –ø–æ—Å—Ç—Ä–æ–µ–Ω–æ. –í—Å–µ —Ä–∞–≤–Ω–æ –∑–∞–≤–µ—Ä—à–∞–µ–º.")
        self.buildings_completed.append(self.current_building)
        print(f"‚úÖ –ó–¥–∞–Ω–∏–µ {self.current_building['name']} –∑–∞–≤–µ—Ä—à–µ–Ω–æ!")
        self.current_building = None
        return self
```

```python

builder = BuilderRobot("–°—Ç—Ä–æ–∏—Ç–µ–ª—å")

print("üìù –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è —Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–∞ –¥–æ–º–∞")
# –°—Ç—Ä–æ–∏–º –¥–æ–º
(builder.start_building("–î–æ–º")
        .set_material("–∫–∏—Ä–ø–∏—á")
        .lay_foundation()
        .build_walls()
        .install_roof()
        .add_windows(4)
        .inspect_quality()
        .if_quality_good()
        .calculate_cost()
        .finish_building())

print("\nüìù –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è —Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–∞ –≥–∞—Ä–∞–∂–∞")
# –°—Ç—Ä–æ–∏–º –≥–∞—Ä–∞–∂
(builder.start_building("–ì–∞—Ä–∞–∂")
        .set_material("–±–µ—Ç–æ–Ω")
        .lay_foundation()
        .build_walls()
        .install_roof()
        .add_windows(1)
        .inspect_quality()
        .if_quality_good() # –ú–æ–∂–µ—Ç –±—ã—Ç—å –Ω–µ –æ—á–µ–Ω—å —Ö–æ—Ä–æ—à–µ–≥–æ –∫–∞—á–µ—Å—Ç–≤–∞
        .calculate_cost()
        .finish_building())

print("\nüìù –î–µ–º–æ–Ω—Å—Ç—Ä–∞—Ü–∏—è —Å—Ç—Ä–æ–∏—Ç–µ–ª—å—Å—Ç–≤–∞ –º–Ω–æ–≥–æ—ç—Ç–∞–∂–∫–∏")
# –°—Ç—Ä–æ–∏–º –º–Ω–æ–≥–æ—ç—Ç–∞–∂–∫—É
(builder.start_building("–ú–Ω–æ–≥–æ—ç—Ç–∞–∂–∫–∞")
        .add_floor("–ø–æ–¥–≤–∞–ª")
        .add_floor("–ø–µ—Ä–≤—ã–π")
        .add_floor("–≤—Ç–æ—Ä–æ–π")
        .add_room("–∫—É—Ö–Ω—è", 1)
        .add_room("—Å–ø–∞–ª—å–Ω—è", 2)
        .set_material("–∂–µ–ª–µ–∑–æ–±–µ—Ç–æ–Ω")
        .lay_foundation()
        .build_walls()
        .install_roof()
        .add_windows(10) # –ú–Ω–æ–≥–æ –æ–∫–æ–Ω –≤ –º–Ω–æ–≥–æ—ç—Ç–∞–∂–∫–µ
        .inspect_quality()
        .if_quality_good()
        .calculate_cost()
        .finish_building())

print(f"\nüìä –ò–¢–û–ì:")
print(f"–ü–æ—Å—Ç—Ä–æ–µ–Ω–æ –∑–¥–∞–Ω–∏–π: {len(builder.buildings_completed)}")
print(f"–û–±—â–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –≤—Å–µ—Ö –∑–¥–∞–Ω–∏–π: {builder.total_cost}")

# –ü—Ä–æ–≤–µ—Ä–∫–∞, —á—Ç–æ –∑–¥–∞–Ω–∏—è –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –¥–æ–±–∞–≤–ª–µ–Ω—ã
if len(builder.buildings_completed) >= 2:
    print("‚úÖ –¢–ï–°–¢ –ü–†–û–ô–î–ï–ù: –£—Å–ø–µ—à–Ω–æ –ø–æ—Å—Ç—Ä–æ–µ–Ω–æ –Ω–µ—Å–∫–æ–ª—å–∫–æ –∑–¥–∞–Ω–∏–π.")
else:
    print("‚ùå –¢–ï–°–¢ –ü–†–û–í–ê–õ–ï–ù: –ù–µ–¥–æ—Å—Ç–∞—Ç–æ—á–Ω–æ –∑–∞–≤–µ—Ä—à–µ–Ω–Ω—ã—Ö –∑–¥–∞–Ω–∏–π.")

if builder.total_cost > 0:
    print("‚úÖ –¢–ï–°–¢ –ü–†–û–ô–î–ï–ù: –û–±—â–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–∞.")
else:
    print("‚ùå –¢–ï–°–¢ –ü–†–û–í–ê–õ–ï–ù: –û–±—â–∞—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –Ω–µ —Ä–∞—Å—Å—á–∏—Ç–∞–Ω–∞.")
```
